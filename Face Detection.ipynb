```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
eye_cascade = cv2.CascadeClassifier("eye.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    single_face = img[y:y+h,x:x+w]
    eyes = eye_cascade.detectMultiScale(single_face,1.22)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
plt.imshow(img)
plt.show()
```

```python
import cv2
cam = cv2.VideoCapture(0)
while True:
    ret,img = cam.read()
    img = cv2.flip(img,1)
    cv2.imshow("WebCam",img)
    if cv2.waitKey(1) == 13:    ## 13 - ASCII Code for Enter
        break
cam.release()
cv2.destroyAllWindows()
```

```python
import cv2
cam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("face.xml")
eye_cascade = cv2.CascadeClassifier("eye.xml")
while True:
    ret,img = cam.read()
    img = cv2.flip(img,1)
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        single_face = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(single_face,1.22)
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
    cv2.imshow("WebCam",img)
    if cv2.waitKey(1) == 13:    ## 13 - ASCII Code for Enter
        break
cam.release()
cv2.destroyAllWindows()
```

```python
### Detect the object by color(red)
#### Access a Webcam
import cv2
import numpy as np
cam = cv2.VideoCapture(0)   ## 0 - default webcam
while True:
    ret,img = cam.read()
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    low = np.array([140,150,0])
    high = np.array([180,255,255])
    img_mask = cv2.inRange(hsv,low,high)
    final_image = cv2.bitwise_and(img,img,mask=img_mask)
    cv2.imshow("Final Image",final_image)
    cv2.imshow("Mask Image",img_mask)
    cv2.imshow("HSV",hsv)
    cv2.imshow("Webcam",img)
    if cv2.waitKey(1) == 13:     ### 13 - ASCII Code for enter
        break
cam.release()
cv2.destroyAllWindows()
```

```python
### Detect Object By Colour(Blue)
import cv2
import numpy as np
cam = cv2.VideoCapture(0)
while True:
    ret,img = cam.read()
    img = cv2.flip(img,1)
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    ## Blue Color
    low = np.array([[100,50,50]])
    high = np.array([[140,255,255]])
    mask_image = cv2.inRange(hsv,low,high)
    output = cv2.bitwise_and(img,img,mask=mask_image)
    cv2.imshow("Mask Image",mask_image)
    cv2.imshow("WebCam",img)
    cv2.imshow("HSV",hsv)
    cv2.imshow("Final Image",output)
    if cv2.waitKey(1) == 13:
        break
cam.release()
cv2.destroyAllWindows()
```

```python
### Object Detection by Color (Green Colour)
import cv2
import numpy as np
cam = cv2.VideoCapture(0)
while True:
    ret,img = cam.read()
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    #Green
    low = np.array([40,50,50])
    high = np.array([140,255,255])
    mask_image = cv2.inRange(hsv,low,high)
    final_image = cv2.bitwise_and(img,img,mask=mask_image)
    cv2.imshow("Final Image",final_image)
    cv2.imshow("Mask Image",mask_image)
    cv2.imshow("HSV",hsv)  ### hue saturation value
    cv2.imshow("WebCam",img)
    if cv2.waitKey(1) == 13:          ### ascii code for enter
        break
cam.release()
cv2.destroyAllWindows()
```
