```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.amazon.in/Samsung-Galaxy-Electric-Blue-128GB-Storage/product-reviews/B085J1J32G/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
names = soup.select("span.a-profile-name")[2:]
names
```

```python
title = soup.select('a.review-title span')
title
```

```python
stars = soup.select("span.a-icon-alt")[3:]
stars
```

```python
dates = soup.select('span.review-date')[2:]
dates
```

```python
reviews = soup.select('span.review-text span')
reviews
```

```python
cust_name = []
rev_title = []
rev_date = []
ratings = []
rev_detail = []
for i in range(len(names)):
    cust_name.append(names[i].get_text())
    rev_title.append(title[i].get_text())
    rev_date.append(dates[i].get_text().replace("Reviewed in India on ",""))
    ratings.append(stars[i].get_text())
    rev_detail.append(reviews[i].get_text().strip("\n"))
```

```python
rev_detail
```

```python
import pandas as pd
df = pd.DataFrame()
df["Date"] = rev_date
df["Name"] = cust_name
df["Ratings"] = ratings
df['Title'] = rev_title
df['Reviews'] = rev_detail
df
```

```python
df.to_csv("reviews.csv")
```
