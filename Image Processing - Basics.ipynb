### Drawing different shapes on image

```python
### Blank Image (512x512)
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
plt.imshow(img)
plt.show()
```

```python
### Draw a line
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.line(img,(50,200),(300,30),(255, 0, 0),5)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.rectangle(img,(50,200),(300,30),(255, 0, 0),5)
plt.imshow(img)
plt.show()
```

```python
### Draw a rectangle
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.rectangle(img,(50,200),(300,30),(255, 0, 0),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw a Circle
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.circle(img,(250,250),200,(255, 0, 0),5)
cv2.circle(img,(250,250),10,(0, 0, 255),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw an Ellipse
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.ellipse(img,(250,250),(200,50),0,0,360,(255, 0, 0),5)
plt.imshow(img)
plt.show()
```

```python
### Draw an Polygon
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
points = np.array([[10,200],[100,100],[400,200],[500,90]])
cv2.polylines(img,[points],True,(255, 0, 0),5)
plt.imshow(img)
plt.show()
```

```python
### Write a text
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.putText(img,"ARPIT",(100,200),cv2.FONT_HERSHEY_TRIPLEX,2.5,(255, 0, 0),3)
plt.imshow(img)
plt.show()
```

```python
### Write a text
import numpy as np
img = np.zeros((512,512,3),np.uint8)+255
cv2.putText(img,"ARPIT",(100,200),cv2.FONT_HERSHEY_TRIPLEX,2.5,(255, 0, 0),3)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.imwrite("image.jpg",img)
plt.imshow(img)
plt.show()
```
